### Plot Types

#### What were the different sports in the first olympics ? Plot them using different graphs

```python
foDF=athletesDF[athletesDF.Edition==1896]
foDF.head()
```

```python
foDF.Sport.value_counts()
```

### Line Plot

```python
foDF.Sport.value_counts().plot(kind='line')
```

```python
foDF.Sport.value_counts().plot(kind='line');
```

### Bar Plot

```python
foDF.Sport.value_counts().plot(kind='bar');
```

### Horizontal Bar Plot

```python
foDF.Sport.value_counts().plot(kind='barh');
```

### Pie Chart

```python
foDF.Sport.value_counts().plot(kind='pie');
```

```python
foDF.Sport.value_counts().plot(color='plum');
```

```python
foDF.Sport.value_counts().plot(kind='bar',color='maroon');
```

```python
foDF.Sport.value_counts().plot(figsize=(10,3));
```

```python
foDF.Sport.value_counts().plot(kind='pie',colormap='Pastel1');
```
